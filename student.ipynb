# Student Information Collection System

This notebook collects and manages student information including name, age, gender, ID, and marks.

## 2. Define Student Input Function

```python
def get_student_input():
    """
    Function to collect student information from user input.
    Returns a dictionary with student details.
    """
    try:
        name = input("Enter student name: ").strip()
        if not name:
            print("Name cannot be empty!")
            return None
        
        age = int(input("Enter student age: "))
        if age < 5 or age > 100:
            print("Age must be between 5 and 100!")
            return None
        
        gender = input("Enter student gender (Male/Female/Other): ").strip().capitalize()
        if gender not in ["Male", "Female", "Other"]:
            print("Please enter valid gender (Male/Female/Other)")
            return None
        
        student_id = input("Enter student ID: ").strip()
        if not student_id:
            print("Student ID cannot be empty!")
            return None
        
        marks = float(input("Enter student marks (0-100): "))
        if marks < 0 or marks > 100:
            print("Marks must be between 0 and 100!")
            return None
        
        student_data = {
            "Name": name,
            "Age": age,
            "Gender": gender,
            "Student ID": student_id,
            "Marks": marks
        }
        return student_data
    
    except ValueError:
        print("Invalid input! Please enter correct data types.")
        return None
```

## 3. Collect Student Information

```python
# Initialize list to store student records
students = []

# Collect information for multiple students
while True:
    print("\n--- Enter Student Information ---")
    student = get_student_input()
    
    if student is not None:
        students.append(student)
        print("✓ Student information added successfully!\n")
    
    another = input("Do you want to add another student? (yes/no): ").strip().lower()
    if another != 'yes' and another != 'y':
        break

print(f"\nTotal students added: {len(students)}")
```

## 4. Display Student Data

```python
# Display collected student information
if students:
    print("\n" + "="*70)
    print("STUDENT INFORMATION COLLECTED")
    print("="*70)
    
    for i, student in enumerate(students, 1):
        print(f"\nStudent {i}:")
        print(f"  Name: {student['Name']}")
        print(f"  Age: {student['Age']}")
        print(f"  Gender: {student['Gender']}")
        print(f"  Student ID: {student['Student ID']}")
        print(f"  Marks: {student['Marks']}")
    print("\n" + "="*70)
else:
    print("No student information to display.")
```

## 5. Store Data in a Data Structure

```python
# Convert to pandas DataFrame
if students:
    df = pd.DataFrame(students)
    print("\nStudent Data as DataFrame:")
    print(df)
    
    # Calculate statistics
    print("\n" + "="*70)
    print("STATISTICS")
    print("="*70)
    print(f"Total Students: {len(df)}")
    print(f"Average Age: {df['Age'].mean():.2f}")
    print(f"Average Marks: {df['Marks'].mean():.2f}")
    print(f"Highest Marks: {df['Marks'].max()}")
    print(f"Lowest Marks: {df['Marks'].min()}")
    
    # Option to save to CSV
    save_option = input("\nDo you want to save the data to a CSV file? (yes/no): ").strip().lower()
    if save_option in ['yes', 'y']:
        filename = "student_data.csv"
        df.to_csv(filename, index=False)
        print(f"✓ Data saved to {filename}")
else:
    print("No student data to save.")
```
